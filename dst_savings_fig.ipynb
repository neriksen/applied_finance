```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Description of data/disposible_income_for_fig.csv', sep=';', index_col=0)
```

```python
plt.style.use('seaborn')
fig, ax = plt.subplots(1, 1, figsize=(7, 6))
sns.lineplot(data = data, linewidth = 4, linestyle="solid")
ax.set_xlabel('Age')
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set_ylabel('DKK', rotation=0, loc='top', labelpad=-21)
ax.legend(['Interpolated median disposible income', 'Polynomial fit', 'Polynomial fit 3% y/y slope increase'])
plt.tight_layout()
fig.savefig('Description of data/disposable_income.png')
```
