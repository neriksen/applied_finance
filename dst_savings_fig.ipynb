```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 50)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.max_rows', 1000)
plt.style.use('seaborn')
plt.rc('font', size=10)          # controls default text sizes
plt.rc('axes', titlesize=16)     # fontsize of the axes title
plt.rc('axes', labelsize=16)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=13)    # fontsize of the tick labels
plt.rc('ytick', labelsize=13)    # fontsize of the tick labels
plt.rc('legend', fontsize=13)    # legend fontsize
plt.rc('figure', titlesize=13)  # fontsize of the figure title
```

```python
data = pd.read_csv('Description of data/disposable_income_for_fig.csv', sep=';', index_col=0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 8))
sns.lineplot(data = data, linewidth = 4, linestyle="solid")
ax.set_xlabel('Age')
ax.get_yaxis().set_major_formatter(
    matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set_ylabel('DKK', rotation=0, loc='top', labelpad=-79)
ax.legend(['Interpolated median disposible income', 'Polynomial fit', 'Polynomial fit 3% y/y slope increase'])
fig.savefig('Description of data/disposable_income.png', bbox_inches = 'tight', pad_inches = 0.1, dpi=192)
```
